```python
from xraydb import material_mu
import numpy as np

def background_si(energies_eV, fluxes, materials, thicknesses_m):
    """
    Calculate total attenuated background counts from multiple photon energies,
    using thickness in meters.

    Parameters:
        energies_eV (list): Photon energies in eV.
        fluxes (list): Relative fluxes for each energy (same length as energies).
        materials (list): List of absorbing materials (e.g., 'SiO2', 'Au').
        thicknesses_m (list): Thickness of each material in meters.

    Returns:
        total_counts (float): Total relative background counts after attenuation.
        breakdown (list): Attenuated counts for each energy.
    """
    if len(energies_eV) != len(fluxes):
        raise ValueError("Energies and fluxes must be the same length.")
    if len(materials) != len(thicknesses_m):
        raise ValueError("Materials and thicknesses must match.")

    # Convert thickness to cm (since material_mu uses cm⁻¹)
    thicknesses_cm = [t * 100 for t in thicknesses_m]

    attenuated_fluxes = []
    for E, flux in zip(energies_eV, fluxes):
        total_attenuation = 0.0
        for mat, d_cm in zip(materials, thicknesses_cm):
            mu = material_mu(mat, E)  # linear attenuation coeff in cm⁻¹
            total_attenuation += mu * d_cm
        transmitted = flux * np.exp(-total_attenuation)
        attenuated_fluxes.append(transmitted)

    total_counts = sum(attenuated_fluxes)
    return total_counts, attenuated_fluxes
```

```python
# Define compound_mass_attn function
def compound_mass_attn(elements: dict, energy: float) -> float:
    """Compute mass attenuation coefficient (cm²/g) for a compound."""
    mass_sum = sum(val * xraydb.atomic_mass(el) for el, val in elements.items())
    mu_mass = sum(xraydb.mu_elam(el, energy) * val * xraydb.atomic_mass(el) / mass_sum
                  for el, val in elements.items())
    return mu_mass


def solve_wls_system(A, y, dy):
    """
    Perform a weighted least squares fit with uncertainty propagation.

    Parameters:
        A  : np.ndarray  — design matrix (N×M)
        y  : np.ndarray  — measured data (N,)
        dy : np.ndarray  — uncertainties on y (N,)

    Returns:
        x  : np.ndarray  — best-fit parameters (M,)
        dx : np.ndarray  — 1σ uncertainties on parameters (M,)
    """
    W = np.diag(1 / dy**2)
    Aw = W @ A
    yw = W @ y

    x = np.linalg.lstsq(Aw, yw, rcond=None)[0]
    residuals = y - A @ x
    norm_residuals = residuals / dy
    dof = len(y) - len(x)
    reduced_chi2 = np.sum(norm_residuals**2) / dof
    cov_x = np.linalg.inv(A.T @ W @ A) * reduced_chi2
    dx = np.sqrt(np.diag(cov_x))

    y_fit = A @ x
    dy_fit = np.sqrt(np.sum(A @ cov_x * A, axis=1))

    return x, dx, y_fit, dy_fit




# Define photon energies
energies_keV = [6.5e3, 14.4e3, 122.0e3]

# Define measurement setups [m], if cardboard is given in g/m²
to_cm = 1e2
to_gcm2 = 1e-4
cardboard_elements = {'C': 6, 'H': 10, 'O': 5}
thickness_gm2_cardboard = 600

setups_real = [
    {"Al":0.0,         "Fe":0.0,         "Quartz":0.0,        "Cardboard":0.0},
    {"Al":25e-6,       "Fe":0.0,         "Quartz":0.0,        "Cardboard":0.0},
    {"Al":50e-6,       "Fe":0.0,         "Quartz":0.0,        "Cardboard":thickness_gm2_cardboard},
    {"Al":0.0,         "Fe":0.0,         "Quartz":0.0,        "Cardboard":thickness_gm2_cardboard},
    {"Al":0.0,         "Fe":10e-6,       "Quartz":0.0,        "Cardboard":thickness_gm2_cardboard},
    {"Al":0.0,         "Fe":20e-6,       "Quartz":0.0,        "Cardboard":thickness_gm2_cardboard*2},
    {"Al":0.0,         "Fe":25e-6,       "Quartz":0.0,        "Cardboard":0.0},
    {"Al":0.0,         "Fe":25e-6,       "Quartz":0.5e-3,     "Cardboard":0.0},
    {"Al":0.0,         "Fe":100e-6,      "Quartz":0.0,        "Cardboard":thickness_gm2_cardboard},
]


# Measured rates and uncertainties
measured_rates = np.array([42000, 38000, 30000, 34000, 25000, 15000, 14000, 11000, 8000])
uncertainties = np.sqrt(measured_rates)



# Compute transmission matrix 

T_matrix = []

for setup in setups_real:
    row = []
    for energy in energies_keV:
        attenuation = 0.0
        for material, thickness in setup.items():
            if thickness == 0:
                mu = 0
            if material == "Cardboard":
                mu_mass = compound_mass_attn(cardboard_elements, energy)
                mu = mu_mass * thickness * to_gcm2  # thickness is in g/cm²
            else:
                mu = xraydb.material_mu(material, energy) * thickness * to_cm  # thickness in cm
                #print(thickness * to_cm)
            attenuation += mu
            #print(np.exp(-attenuation))
        row.append(np.exp(-attenuation))
        #print(np.exp(-attenuation))
    T_matrix.append(row)

T_matrix = np.array(T_matrix)


x, dx, y_fit, dy_fit = solve_wls_system(T_matrix, measured_rates, uncertainties)
x,dx
```
